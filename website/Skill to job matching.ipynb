```python
def get_matches(entered_cl = entered_cl):
    
    
    scores = [] # what part of requirments is satisfied
    count = 0
    for i in x:
        inters = list(set(entered_cl) & set(i))
        score = 0

        for j in inters:
            score+=len(i) - i.index(j)
        if len(i)!=0:
            scores.append( [2*score/(len(i)*(len(i)+1)), count])
        else:
            scores.append([-1, count])
        count+=1
        
    scores.sort(reverse = True)
    most_def_get=[scores[i][1] for i in range(10)]
    
    
    # here we wight scores with number of intesections , so it will generate more challenging ones
    intersects = get_interects(entered_cl)
    good_match = [[scores[i][0]*len(intersects[scores[i][1]]),scores[i][1]] for i in range(len(scores)) ]
    good_match.sort(reverse=True)
    harder_to_get = [good_match[i][1] for i in range(10)]
    
    return most_def_get, harder_to_get, intersects 
```

```python
most_def_get, harder_to_get, intersects = get_matches()



best = (set(most_def_get) & set(harder_to_get))
inters = [intersects[i] for i in best]


print ('the best vacancies are numbers '+ str(best))
print('the skills that match for each are ' + str(inters))
print ('the skills that miss foe each are ' + str([set(x[i])-set(intersects[i]) for i in best]))
# ofc we also need a text on website
    
print('more difficult ones are '+str (set(harder_to_get) - (best)))
print('more easy ones are '+str (set(most_def_get) - (best)))
```
